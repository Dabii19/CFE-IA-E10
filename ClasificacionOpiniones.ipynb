```python
#Se importan pandas y se muestras los primeros 5

import pandas as pd

df = pd.read_csv("dataset_train_95.csv")

df.head(5)
```

```python
#Se observan si hay datos vacías

df.info()
```

```python
#Observa los nulos

df.isnull().sum()
```

```python
#FILAS DUPLICADAS ANTES DE ELIMINAR

print("Filas duplicadas antes de eliminar", df.shape)

#Elimina duplicados
df.drop_duplicates(inplace = True)

print("Filas duplicadas después de eliminar", df.shape)
```

```python
#Creacion del modelo

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['declaracion'], df['etiqueta'], test_size=0.10)

from sklearn.model_selection import train_test_split

X2_train, X2_test, y2_train, y2_test = train_test_split(df['tweet'], df['etiqueta'], test_size=0.10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizador = CountVectorizer()
X_train_vec = vectorizador.fit_transform(X_train)
X_test_vec = vectorizador.transform(X_test)

from sklearn.feature_extraction.text import CountVectorizer

vectorizador = CountVectorizer()
X2_train_vec = vectorizador.fit_transform(X2_train)
X2_test_vec = vectorizador.transform(X2_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

modelo = LogisticRegression()
modelo.fit(X_train_vec, y_train)

predit = modelo.predict(X_test_vec)

accuracy = accuracy_score(y_test, predit)
print("Porcentaje de acierto primer modelo:", accuracy * 100)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

modelo = LogisticRegression()
modelo.fit(X2_train_vec, y2_train)

predit2 = modelo.predict(X2_test_vec)

accuracy2 = accuracy_score(y2_test, predit2)
print("Porcentaje de acierto segundo modelo:", accuracy2 * 100)
```
